# Network Data Science with NetworkX and Python

## Generate random graphs

```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
graph = nx.complete_graph(10)
```

```python
fig = plt.figure(figsize = (14, 9))
nx.draw(graph, 
        node_color = 'C1',
        with_labels = True,
        edge_color = 'g')
fig.set_facecolor('k')
```

```python
fig = plt.figure(figsize = (14, 9))
graph = nx.binomial_tree(5)
nx.draw(graph, 
        node_color = 'C1',
        with_labels = True,
        edge_color = 'g')
fig.set_facecolor('k')
```

```python
fig = plt.figure(figsize = (14, 9))
graph = nx.binomial_graph(20, 0.15)
nx.draw(graph, 
        node_color = 'C1',
        with_labels = True,
        edge_color = 'g')
fig.set_facecolor('k')
```

```python
fig = plt.figure(figsize = (14, 9))
graph = nx.connected_caveman_graph(5, 10)
nx.draw(graph, 
        node_color = 'C1',
        with_labels = True,
        edge_color = 'g')
fig.set_facecolor('k')
```

```python
fig = plt.figure(figsize = (14, 9))
graph = nx.random_internet_as_graph(1000)
nx.draw(graph, 
        node_color = 'C1',
        with_labels = True,
        edge_color = 'g')
fig.set_facecolor('k')
```

```python
fig = plt.figure(figsize = (14, 9))
graph = nx.les_miserables_graph()
nx.draw(graph, 
        node_color = 'C1',
        with_labels = True,
        edge_color = 'g')
fig.set_facecolor('c')
```
